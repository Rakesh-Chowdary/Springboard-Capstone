## Let us pull these four columns from all the datasets and store into a single large Data Frame

### Columns:
    1. product_category
    2. star_rating
    3. review_headline
    4. review_body

```python
%%capture capt
chunk_list = []  # append each chunk df here 

for each_file_name in list_files:
    
    print("Intitating reading data from file: {}".format(each_file_name))
    
    df_chunk = pd.read_csv("Reviews Data/"+each_file_name,delimiter='\t',encoding='utf-8',error_bad_lines=False,header=0, chunksize=500000)
    
    # Each chunk is in df format
    for chunk in df_chunk:  
        # perform data filtering 
        #chunk_filter = chunk_preprocessing(chunk)
        
        # Once the data filtering is done, append the chunk to list
        chunk_list.append(chunk[['product_category','star_rating','review_headline','review_body']])
        
    print("completed appending data from file: {}".format(each_file_name))
    
# concat the list into dataframe 
main_df = pd.concat(chunk_list)
```
