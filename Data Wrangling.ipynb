<h1><center>Data Wrangling</center></h1>

Now as we downloaded the data let's load the data and do some data wrangling.

```python
import os
```

```python
list_files = []
for each_file in os.listdir(path="Reviews Data"):
    if str(each_file).endswith('tsv.gz'):
        list_files.append(each_file)
```

```python
print(list_files)
```

```python
print("Number of files from Reviews Data: {}".format(len(list_files)))
```

## Now we have all the files respective to the download links from data collection

Let's us load each tar file into pandas and extract the columns we need

```python
import pandas as pd
```

## Let us pull these four columns from all the datasets and store into a single large Data Frame

### Columns:
    1. product_category
    2. star_rating
    3. review_headline
    4. review_body

```python
main_DataFrame = pd.DataFrame([])
```

```python
%%capture
for each_file_name in list_files:
    print("Intitating reading data from file: {}".format(each_file_name))
    main_DataFrame.append((pd.read_csv("Reviews Data/"+str(each_file_name),\
                                       delimiter='\t',\
                                       encoding='utf-8',\
                                       error_bad_lines=False,\
                                       header=0))\
                          [['product_category','star_rating','review_headline','review_body']])###
    
    print("completed appending data from file: {}".format(each_file_name))
```
